```python
import numpy as np
import matplotlib.pyplot as plt
from re import sub
from itertools import groupby
```

```python
def encode(text):
    return sub(r'(.)\1*', lambda m: "{0:b}".format(len(m.group(0))),
               text)

encode('011101011')
```

```python
def rlencode(input_string):
    return [(len(list(g))) for k,g in groupby(input_string)]

rlencode('011101011')
```

```python
def count_ones(m):
    return np.count_nonzero(m == 1)
```

```python
x = []
y = []
for i in range(0, 1000000):
    probability = i / 1000000.0
    m = np.random.choice([0, 1], size=(10,10), p=[probability, 1 - probability])
    l = ''.join([str(t) for p in m for t in p])
    y.append(len(rlencode(l)))
    x.append(probability)
#print(x)

plt.xlabel('Length of Run-Length-Encoded Data')
plt.ylabel("Percentage of 1's in binary data")
plt.title("RLE of 10x10 Matrices with varying probabilities of '1'")
plt.scatter(y, x, marker='o');
```

```python
x = []
y = []
r = 1000.0
for i in range(0, int(r/2)):
    probability = i / 1000.0
    m = np.random.choice([0, 1], size=(10,10), p=[probability, 1 - probability])
    l = ''.join([str(t) for p in m for t in p])
    y.append(len(encode(str(''.join([str(t) for p in m for t in p])))))
    x.append(probability)
#print(x)
```

```python
plt.xlabel('Length of Run-Length-Encoded Data')
plt.ylabel("Percentage of 1's in binary data")
plt.title("RLE of 10x10 Matrices with varying probabilities of '1'")
plt.scatter(y, x, marker='o');

z4 = np.polyfit(x, y, 3) 
p4 = np.poly1d(z4) # construct the polynomial
plt.plot(p4(x), x, color='red', linewidth=3)
```
